## Celda 5 - Crear hojas

```python
HOJAS={
'DOCUMENTOS':['DOC_ID','TITULO','PAIS','REGULADOR','ORGANIZACION','IDIOMA','TIPO_DOCUMENTO','FECHA','URL','RESUMEN','PALABRAS_CLAVE','RIESGOS','SALUD_MENTAL','VIOLENCIA_DIGITAL','DESINFORMACION','IA','AMI','ESTRATEGIAS','COOPERACION','NIVEL_EVIDENCIA','ESTADO','FECHA_CARGA'],
'RIESGOS':['ID','CATEGORIA','RIESGO','DESCRIPCION','AMI'],
'AMI':['ID','COMPETENCIA','DESCRIPCION','RECURSO'],
'REGULADORES':['ID','PAIS','NOMBRE','TIPO','WEB'],
'PAISES':['ID','PAIS','ISO','REGION'],
'ORGANIZACIONES':['ID','NOMBRE','TIPO','WEB'],
'COOPERACION':['ID','ORGANIZACION','CONVOCATORIA','PAIS','URL','ESTADO']
}
exist=[w.title for w in sh.worksheets()]
for hoja,cols in HOJAS.items():
    if hoja not in exist:
        ws=sh.add_worksheet(title=hoja,rows=2000,cols=max(30,len(cols)))
    else:
        ws=sh.worksheet(hoja)
    if not ws.row_values(1):
        ws.append_row(cols)
    print('✓',hoja)
```
